### Pregunta 6

### Pregunta 8
Utiliza el código para obtener la siguiente información (ubicada en `df`).

- Promedio de los precios de venta de los modelos marca **Realme**
- Máximo de estrellas de los modelos marca **Redmi**
- Mínimo de Estrellas de los modelos marca **Redmi**

```python
Prom = df[df['Nombre Marca'] == 'Realme']['Precio de venta'].mean()
MaxSt = df[df['Nombre Marca'] == 'Realme']['Estrellas'].max()
MinSt = df[df['Nombre Marca'] == 'Realme']['Estrellas'].min()
print(f'La marca Redmi cuenta con equipos con calificación desde {MinSt} hasta {MaxSt} estrellas a precios alrededor de {Prom}')
```

### Pregunta 9
Utiliza un **Groupby** para mostrar la cantidad de estrellas promedio que cada marca tiene en `df`.

```python
BrandSt = df.groupby('Nombre Marca')['Estrellas'].mean().sort_values(ascending = False).reset_index()
BrandSt
```

```python
# Pero esto puede apreciarse muchísimo mejor en un gráfico de barras
plt.figure(figsize = (10,10))
bar_plot = sns.barplot(x = BrandSt['Nombre Marca'], y = BrandSt['Estrellas'])
BrandSt['Estrellas'] = round(BrandSt['Estrellas'],2)
for i, brand in enumerate(BrandSt['Estrellas'].values):
  bar_plot.text(i, brand, str(brand), ha = 'center', va = 'bottom')
plt.show()
```

### Pregunta 10
¿Qué otro dato puedes obtener de `df` y cómo lo extraerías?

#### Calificaciones y estrellas promedio por marca

```python
# Con una gráfica de barras podemos visualizar la cantidad de calificaciones, proporcional al número de compradores de cada marca.

plt.figure(figsize = (20,15))
df['Miles_Calificaciones'] = df['Calificaciones']/1000
BrandCal = df.groupby('Nombre Marca')['Miles_Calificaciones'].sum().reset_index()
BrandCal = BrandCal[BrandCal['Miles_Calificaciones'] >= 150]
BrandCal['Miles_Calificaciones'] = round(BrandCal['Miles_Calificaciones'], 2)
bar_plot = sns.barplot(data = BrandCal, x = 'Nombre Marca', y = 'Miles_Calificaciones', errorbar = None)
plt.xlabel('Marcas')
plt.ylabel('Miles de Calificaciones')
plt.title('Calificaciones y estrellas promedio por marca')
# Agregamos cantidad numérica de calificaciones y promedio de estrellas por marca
for i, val in enumerate(BrandCal['Miles_Calificaciones'].values):
  bar_plot.text(i, val, 'Calificaciones: ' + str(val), ha = 'center', va = 'bottom', fontsize = 9)
  bar_plot.text(i, val, 'Estrellas: ' + str(BrandSt['Estrellas'].values[i]), ha = 'center', va = 'top')
plt.show()

## Mejorar ##
```

##### Conclusiones
- Podemos apreciar una mayor cantidad de calificaciones en equipos de la marca Realme y Redmi, frente al resto. Esto puede significar que los usuarios son más leales o que simplemente éstas dos tienen una mejor forma de llevar esta encuesta a cabo.
-En el gráfico se agregó la cantidad de estrellas promedio por marca, para apreciar la calidad de las calificaciones en función de la cantidad.

#### Matriz de correlación

```python
# Podemos visualizar si existe correlación entre las columnas
df_imp = ['RAM_GB','ROM_GB','Calificaciones','Rese–as','Estrellas','Precio de lista','Precio de venta']
df_imp = df[df_imp]
plt.figure(figsize = (10,10))
sns.heatmap(data = df_imp.corr(), annot = True)
plt.title('Matriz de correlación')
plt.show()
```
